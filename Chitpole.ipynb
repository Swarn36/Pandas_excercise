##How many items were orderd in total?

```python
total_orders = chipo.quantity.sum()
total_orders
```

##Turn the item price into a float

```python
chipo.item_price.dtype
```

```python
dollarizer = lambda x: float(x[1:-1])
chipo.item_price = chipo.item_price.apply(dollarizer)
chipo.item_price.dtype
```

##How much was the revenue for the period in the dataset?

```python
chipo['revenue'] = chipo['item_price'] * chipo['quantity']
chipo.head()
```

```python
total_revenue = chipo.revenue.sum()
print(f'Total revenue for the period is ${total_revenue}')
```

##How many orders were made in the period?

```python
c_oi = chipo.order_id.value_counts()
total_orders = c_oi.count()
print(f'Total orders were {total_orders}')
```

```python
chipo.groupby(by = 'order_id').sum()
```

##What is the average revenue amount per order?

```python
average_revenue_per_order = total_revenue / total_orders
print(f'Revenue per order ${str(round(average_revenue_per_order,2))}')
```

```python
#another way to do it
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
order_grouped = chipo.groupby(by=['order_id']).sum()
order_grouped.mean()['revenue']
```

```python
order_grouped['revenue'].mean()
```
