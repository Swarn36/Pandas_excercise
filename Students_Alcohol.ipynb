##For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column

```python
stud_alcho = df.loc[:, 'school':'guardian']
stud_alcho.head()
```

Create a lambda function that will capitalize strings.

```python
capitalizer = lambda x: x.capitalize()
```

Capitalize both Mjob and Fjob

```python
stud_alcho['Mjob'] = stud_alcho['Mjob'].apply(capitalizer)
stud_alcho['Fjob'] = stud_alcho['Fjob'].apply(capitalizer)
stud_alcho.head()
```

Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def majority(x):
  if x > 17:
    return 1
  else:
    return 0

stud_alcho['legal_drinker'] = stud_alcho['age'].apply(majority)
stud_alcho.head()
```

```python
d = {1: 'legal', 0: 'illegal'}
stud_alcho['legal_drinker'] = stud_alcho['legal_drinker'].map(d)
stud_alcho.head()
```

Multiply every number of the dataset by 10.

```python
def multiple10(x):
  if type(x) is int:
    return x*10
  return x

stud_alcho.applymap(multiple10)
```
